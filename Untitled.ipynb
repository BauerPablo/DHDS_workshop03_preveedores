```python
def  mag_identify(desc):
    if 'TORQUIMETRO'in desc or 'TORQUÍMETRO' in desc:
        desc = 'TORQUE'
    elif 'DINAM' in desc or 'CALIBRE' in desc or 'NIVEL' in desc or 'PENETRADOR' in desc or 'MICROMETRO'in desc or 'MICRÓMETRO' in desc or 'MANDRILADORA' in desc or 'SONDAS' in desc or 'DUROMETRO' in desc or 'PARALELA' in desc or 'BALANZA' in desc or 'COMPARADOR' in desc or 'CINTA' in desc or 'DUROMETRO' in desc or 'CRIMPING' in desc or 'DUREZA' in desc:
        desc = 'DIMENSIONAL'
    elif 'MANOMETRO'in desc or 'MANÓMETRO' in desc or 'VACU' in desc or 'PRESSURE' in desc:
        desc = 'PRESIÓN'
    elif 'TEMP' in desc or 'TERMO' in desc or 'HORNO' in desc:
        desc = 'TEMPERATURA'
    elif 'MEGOHM' in desc or 'ELECTRICIDAD' in desc or 'MULTIM' in desc or 'FUENTE' in desc or 'VOLTI' in desc or 'GAUSS' in desc or 'OHMMETER' in desc or 'AMPER' in desc or 'RESIS' in desc or 'CELDA' in desc or 'CRONO' in desc:
        desc = 'ELECTRICIDAD'
    elif 'BORO' in desc or 'EDDY' in desc or 'ULTRASON' in desc:
        desc = 'NDT'
    else:
        desc = 'OTROS'
    return desc
```

```python
ds_complete['magnitud'] = ds_complete['description'].apply(lambda x: mag_identify(x))
```

```python
def tiempos(mag):
    if mag == 'DIMENSIONAL':
        time = 5
    elif mag == 'PRESIÓN':
        time = 4
    elif mag == 'TORQUE':
        time = 4
    elif mag == 'OTROS':
        time = 5
    elif mag == 'TEMPERATURA':
        time = 5
    elif mag == 'ELECTRICIDAD':
        time = 5
    elif mag == 'NDT':
        time = 5
    
    return time
```

```python
ds_complete['tiempo_calibracion'] = ds_complete.magnitud.apply(lambda x: tiempos(x))
```

```python
ds_complete['costo_calibracion_lab'] = ds_complete.tiempo_calibracion.apply(lambda x: x * 60)
```
